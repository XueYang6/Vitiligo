```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import copy
from Interpretable_machine_learning import InterpretableMachineLearning
from Interpretable_machine_learning import draw_all_roc, draw_confusion_matrix, draw_roc
import os
from tqdm import tqdm
```

# GI

```python
df_GI = pd.read_csv("G:/AIM/work/leucoderma/excel/预处理-0.5-1820-GI-4.csv")
IMLGI = InterpretableMachineLearning(df_GI, 'Non-segmental vitiligo')
x = IMLGI.x 
y = IMLGI.y

acc_rfc_all = []
acc_xgb_all = []  
acc_svm_all = []
acc_gbdt_all = []
auc_rfc_all = []
auc_xgb_all = []
auc_svm_all = []
auc_gbdt_all = []

best_rfc_acc = 0
best_xgb_acc = 0
best_svm_acc = 0
best_gbdt_acc = 0

best_rfc_auc = 0
best_xgb_auc = 0
best_svm_auc = 0
best_gbdt_auc = 0

best_rfc_pre = []
best_xgb_pre = []
best_svm_pre = []
best_gbdt_pre = []

best_rfc_proba = []
best_xgb_proba = []
best_svm_proba = []
best_gbdt_proba = []

for i in tqdm(range(50)):
    y_train_rfc, y_test_rfc, y_pre_rfc, y_proba_rfc, acc_rfc, auc_rfc, rfc_model = IMLGI.model_prediction(x, y, model='rfc', ROS=False)
    y_train_xgb, y_test_xgb, y_pre_xgb, y_proba_xgb, acc_xgb, auc_xgb, xgb_model = IMLGI.model_prediction(x, y, model='xgb', ROS=False)
    y_train_svm, y_test_svm, y_pre_svm, y_proba_svm, acc_svm, auc_svm, svm_model = IMLGI.model_prediction(x, y, model='svm', ROS=False)
    y_train_gbdt, y_test_gbdt, y_pre_gbdt, y_proba_gbdt, acc_gbdt, auc_gbdt, gbdt_model = IMLGI.model_prediction(x, y, model='gbdt', ROS=False)
    acc_rfc_all.append(acc_rfc)
    acc_xgb_all.append(acc_xgb) 
    acc_svm_all.append(acc_svm)
    acc_gbdt_all.append(acc_gbdt)
    auc_rfc_all.append(auc_rfc)
    auc_xgb_all.append(auc_xgb)
    auc_svm_all.append(auc_svm)
    auc_gbdt_all.append(auc_gbdt)
    if acc_rfc > best_rfc_acc:
        best_rfc_pre = y_pre_rfc
        best_rfc_proba = y_proba_rfc
    if acc_xgb > best_xgb_acc:
        best_xgb_pre = y_pre_xgb
        best_xgb_proba = y_proba_xgb
    if acc_svm > best_svm_acc:
        best_svm_pre = y_pre_svm
        best_svm_proba = y_proba_svm
    if acc_gbdt > best_gbdt_acc:
        best_gbdt_pre = y_pre_gbdt
        best_gbdt_proba = y_proba_gbdt
        

gi_acc = [acc_rfc_all, acc_xgb_all, acc_svm_all, acc_gbdt_all]
gi_auc = [auc_rfc_all, auc_xgb_all, auc_svm_all, auc_gbdt_all]

if not os.path.exists('images\GI'):
    os.mkdir('images\GI')
if not os.path.exists('images\GI\CM'):
    os.mkdir('images\GI\CM')
    
draw_confusion_matrix(y_test_rfc, best_rfc_pre,['Segmental vitiligo','Non-segmental vitiligo'], 'GI-RFC-ROS')
plt.savefig('images\GI\CM\gi-rfc.jpg', dpi=2000)
plt.clf()
plt.close()
draw_confusion_matrix(y_test_xgb, best_xgb_pre,['Segmental vitiligo','Non-segmental vitiligo'], 'GI-XGB-ROS')
plt.savefig('images\GI\CM\gi-xgb.jpg', dpi=2000)
plt.clf()
plt.close()
draw_confusion_matrix(y_test_svm, best_svm_pre,['Segmental vitiligo','Non-segmental vitiligo'], 'GI-SVM-ROS')
plt.savefig('images\GI\CM\gi-svm.jpg', dpi=2000)
plt.clf()
plt.close()
draw_confusion_matrix(y_test_gbdt, best_gbdt_pre,['Segmental vitiligo','Non-segmental vitiligo'], 'GI-GBDT-ROS')
plt.savefig('images\GI\CM\gi-gbdt.jpg', dpi=2000)
plt.clf()
plt.close()

draw_all_roc([y_test_rfc, y_test_xgb, y_test_svm, y_test_gbdt], [best_rfc_proba, best_xgb_proba, best_svm_proba, best_gbdt_proba], ['#1d89e4', '#f1569d', '#ff6a25', '#ff0d56'], ['RFC', 'XGB', 'SVM', 'GBDT'], "GI-ROC")
plt.savefig('images\GI\gi-roc.jpg', dpi=2000)
plt.clf()
plt.close()
```
